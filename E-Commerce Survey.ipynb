```python
# import dependencies and set display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# read csv file and drop unnecessary columns
ecom = pd.read_excel('survey_ecom_2019.xlsx')
ecom.head()
```

```python
# Drop unnecessary columns
ecom.drop(ecom.iloc[:, :10]+ecom.iloc[:, 35:37], inplace = True, axis = 1)
ecom.head()
```

```python
# renaming unnamed columns and save questions
questions = {}
count = 0
for column in ecom.columns:
    if "Unnamed" in column:
        pass
    else:
        count += 1
        name = "Q{}".format(count)
        questions[name] = column
    ecom.rename(columns={column:name}, inplace=True)

ecom.head(2)
```

```python
# creating multi-level columns 
level2 = []
level3 = []
for i, choice in enumerate(ecom.iloc[0]):
    if (' - ' in choice) and (i > 30):
        c2 = choice.split(' - ')[0]
        level3.append(choice.split(' - ')[1])
        if ('(') in c2:
            level2.append(c2.split('(')[0])
        else:
            level2.append(c2)
    else:
        if ('(') in choice:
            level2.append(choice.split('(')[0])
            level3.append(choice.split('(')[0])
        else:
            level2.append(choice)
            level3.append(choice)
        
ecom.columns = pd.MultiIndex.from_arrays([ecom.columns, level2, level3], names=('Question','Choice', 'Range'))
ecom.drop([0], inplace=True)
ecom.reset_index(drop=True, inplace=True)
ecom.head(2)
```

```python
# get dataframe from a particular question
key = 'Q1'
print(questions[key])
df = ecom.xs(key, level=0, axis=1)
df = df.droplevel(1, axis=1).count().rename(index={'Laki-Laki': 'Male', 'Perempuan': 'Female' })
df
```

```python
# visualizing the feature
ax = df.plot(kind="pie", figsize=(5,5), fontsize=14, autopct='%1.1f%%')
plt.xticks(rotation=0, horizontalalignment="center")
plt.ylabel("")
```
